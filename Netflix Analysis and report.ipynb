<h1 style="font-family: 'Times New Roman'; color: red;">Distribution of Movie and TV Show Releases by Category</h1>

```python
grouped_data = df.groupby(['Category', 'year'])['Title'].count().reset_index()

# Create the line chart
sns.lineplot(data=grouped_data, x='year', y='Title', hue='Category')

# Set labels and title
plt.xlabel('year')
plt.ylabel('Count')
plt.title('Movies and TV Shows Released Over Time')
```
